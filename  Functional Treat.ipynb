```python
def calculator(a, b, operator):

    if operator == '+':
        result = a + b
    elif operator == '-':
        result = a - b
    elif operator == '*':
        result = a * b
    elif operator == '/':
        if b != 0:
            result = a / b
        else:
            result = "Error! Division by zero."
    else:
        result = "Invalid operator"

    return



# Input from user
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")


# Function call
output = calculator(num1, num2, op)
print("Result:", output)
```

Types of function Arguments
1. *args
2. **kwargs


```python
def order(*args):  # * is the unpacking operator
    print(args)

order(210,150, True, 160)
print(type(Display))
```

```python
def Display(*args):

    for i in args:
        print(i)

Display(5, 2.3, True, 15)
print(type(Display))
```

Function Recursion 

n! = n × (n-1) × (n-2) × ... × 2 × 1

```python
n = int(input("Enter a number: "))
fact = 1

for i in range(1, n + 1):
    fact = fact * i

print("Factorial of", n, "is", fact)
```

```python
# With map()
nums = [11 , 12, 13, 14]
result = list(map(lambda x: x * 2, nums))
print(result)
```
